**HYPERBOLIC MODEL**

Duncan and Chang (1970) based on the works by Kondner (1963) and Kondner and Zelasko (1963) was used to model the soil and concrete interface in the three dimensional finite element analysis.
Clough and Duncan (1969,1971) described the shear stress and displacement curve of the interface using

$\begin{equation*}
\tau=\frac{w_s}{a+bw_s} 
\end{equation*}$                         (4)

in which $\tau$ is shear stress on the contact surface, $w_s$ is the relative tangential displacement between soil and concrete slab, a and b are fitted parameters using the shear and displacement curve obtained from shear tests

$\begin{equation*}
a=\frac{1}{k_{si}},   b=\frac{1}{\tau_u} 
\end{equation*}$ (5)

In the equation, $k_{si}$ and $\tau_u$ are the initial shear stiffness and asymptotic shear stress value obtained from the hyperbolic curve as shown in Fig. 13.
In the model, τu is normally larger than the peak shear stress τf, therefore, Clough and Duncan
(1971) used a failure ratio ($R_f$) to describe the difference:

$\begin{equation*}
\tau_f=R_f\tau_u 
\end{equation*}$  (6)

The initial stiffness $k_{si}$ is expressed as:


$\begin{equation*}
k_{si}=Kp_a \left( \frac{\sigma_n}{p_a} \right)^n 
\end{equation*}$ (7)


in which $\sigma_n$ and $p_a$ are normal stresses on the contact surface and atmospheric air pressure respectively, n and K are the dimensionless numbers to be determined from curve fitting method.
For sand and concrete interface, we can assume zero cohesion at the interface
$\tau=\sigma_n tan\left( \theta \right)$
 
where $\theta$  is the interface friction angle. By substituting Eqs. (3) to (6) into Eq. (7), we obtain
 
$\begin{equation*}
\tau=\frac{w_s}{\frac{1}{Kp_a \left( \frac{\sigma_n}{p_a} \right)^n} + \frac{R_fw_s}{\sigma_ntan(\theta)}}
\end{equation*}$
(8)

By differentiating the above equation with respect to ws, the tangential stiffness of the interface
can be expressed as

$\begin{equation*}
k_{st}= \frac{\partial\tau}{\partial w_s}={Kp_a \left( \frac{\sigma_n}{p_a} \right)^n} {\left(1- \frac{R_f \tau}{\sigma_n tan(\theta)} \right)^2} 
\end{equation*}$  (9)

For a 3D model, the tangential shear stiffness includes two orthogonal components, i.e., $k_{stx}$ and $k_{sty}$. In this analysis, we consider kst = kstx = ksty for simplicity. To obtain the parameters used in the model, the shear displacement curve obtained from the soil and non-ribbed concrete slab was fitted.
